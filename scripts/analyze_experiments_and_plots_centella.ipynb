```python
mono_front_20_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Mono-objective-results-CEC/20-Days/artificial_repetition/RMDDC_PairBasedCrossover_20Days_0.5_1.txt.csv'
mono_front_40_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Mono-objective-results-CEC/40-Days/artificial_repetition/RMDDC_PairBasedCrossover_40Days_0.5_1.txt.csv'
mono_front_60_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Mono-objective-results-CEC/60-Days/artificial_repetition/RMDDC_PairBasedCrossover_60Days_0.5_1.txt.csv'

multi_front_20_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Multi-objective-results-Mathematics/Centella/20-Days/Centella_Sequential_MenuPlanning_Mathematics_MainStudy_MOEAD_MPP_20_Days_sampleFront.front'
multi_front_40_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Multi-objective-results-Mathematics/Centella/40-Days/Centella_Sequential_MenuPlanning_Mathematics_MainStudy_MOEAD_MPP_40_Days_sampleFront.front'
multi_front_60_path = '/home/amarrero/Proyectos/MenuPlanning_MOEAD_Mathematics_2020/results/Multi-objective-results-Mathematics/Centella/60-Days/Centella_Sequential_MenuPlanning_Mathematics_MainStudy_MOEAD_MPP_60_Days_sampleFront.front'
```

```python
mono_front_20_df = pd.read_csv(mono_front_20_path, sep = ' ', header = None)
mono_front_40_df = pd.read_csv(mono_front_40_path, sep = ' ', header = None)
mono_front_60_df = pd.read_csv(mono_front_60_path, sep = ' ', header = None)

multi_front_20_df = pd.read_csv(multi_front_20_path, sep = ' ', header = None)
multi_front_40_df = pd.read_csv(multi_front_40_path, sep = ' ', header = None)
multi_front_60_df = pd.read_csv(multi_front_60_path, sep = ' ', header = None)
```

```python
mono_front_20_df
```

```python
mono_front_20_df = mono_front_20_df.iloc[:, [-2, -1]]
mono_front_20_df
```

```python
mono_front_40_df = mono_front_40_df.iloc[:, [-2, -1]]
mono_front_60_df = mono_front_60_df.iloc[:, [-2, -1]]

multi_front_20_df = multi_front_20_df.iloc[:, [-2, -1]]
multi_front_40_df = multi_front_40_df.iloc[:, [-2, -1]]
multi_front_60_df = multi_front_60_df.iloc[:, [-2, -1]]
```

```python
def rename_cols(dataset):
    dataset.columns = ['Cost', 'Repetition']
    return dataset
```

```python
mono_front_20_df = rename_cols(mono_front_20_df)
mono_front_40_df = rename_cols(mono_front_40_df)
mono_front_60_df = rename_cols(mono_front_60_df)

multi_front_20_df = rename_cols(multi_front_20_df)
multi_front_40_df = rename_cols(multi_front_40_df)
multi_front_60_df = rename_cols(multi_front_60_df)
```

```python
def plot_front(mono_dataset, multi_dataset, title):
    plt.figure(figsize=(10, 8))
    plt.scatter(mono_dataset['Cost'], mono_dataset['Repetition'], c = 'blue', marker = '.', label = 'Memetic Algorithm')
    plt.scatter(multi_dataset['Cost'], multi_dataset['Repetition'], c = 'red', marker = '*', label = 'MOEA/D')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_front(mono_front_20_df, multi_front_20_df, 'Front comparison between Memetic Algorithm and MOEA/D for 20-Days MPP after 1 hour.')
```

```python
plot_front(mono_front_40_df, multi_front_40_df, 'Front comparison between Memetic Algorithm and MOEA/D for 40-Days MPP after 2.5 hours.')
```

```python
plot_front(mono_front_60_df, multi_front_60_df, 'Front comparison between Memetic Algorithm and MOEA/D for 60-Days MPP after 5 hours.')
```
